```python
#Importing required librabries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
data=pd.read_csv('D:/DATA/Downloads/matches.csv')
data.head(10)
```

```python
Data=pd.read_csv('D:/DATA/Downloads/deliveries.csv')
Data.head(10)
```

```python
#Merging data
season_data=data[['id','season','winner']]
complete_data=Data.merge(season_data,how='inner',left_on='match_id',right_on='id')
```

```python
#Umpire 3 has columns with null values
data=data.drop(columns=["umpire3"],axis=1)
data.head(10)
```

```python
wins_per_season=data.groupby("season")["winner"].value_counts()
wins_per_season
```

```python
#Matches played in each season
plt.figure(figsize=(20,10))
sns.countplot('season',data=data,palette="muted")
plt.title("Number of Matches Played in Each Season",fontsize=24)
plt.xlabel('Season',fontsize=18)
plt.ylabel('Matches Played',fontsize=18)
plt.show()
```

```python
#Number of Matches won by each team
plt.figure(figsize=(20,10))
sns.countplot(x='winner',data=data,palette="bright")
plt.title("Number of Matches Played in Each Season",fontsize=24)
plt.xticks(rotation=30)
plt.xlabel('Teams',fontsize=18)
plt.ylabel('Number of Matches Won',fontsize=18)
plt.show()
```

```python
data['win_by']=np.where(data['win_by_runs']>0,'Bat First','Bowl First')
```

```python
Win=data.win_by.value_counts()
labels=np.array(Win.index)
sizes=Win.values
colors=['#FFBF89','#FA3749']
plt.figure(figsize=(20,10))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%',shadow=True,startangle=90)
plt.title('Match Result', fontsize=20)
plt.axis('equal')
plt.show()
```

```python
#Matches won by batting and bowling first
plt.figure(figsize=(20,10))
sns.countplot('season',hue='win_by',data=data,palette="winter")
plt.title("Number of Matches won by batting and bowling first",fontsize=24)
plt.xlabel('Season',fontsize=18)
plt.ylabel('Count',fontsize=18)
plt.show()
```

```python
Toss=data.toss_decision.value_counts()
labels=np.array(Toss.index)
sizes=Toss.values
colors=['#AFDCEC','#4B00B2']
plt.figure(figsize=(15,10))
plt.pie(sizes,labels=labels, colors=colors,
       autopct='%1.1f%%',shadow=True, startangle=90)
plt.title('Toss result', fontsize=26)
plt.axis('equal')
plt.show()
```

```python
#Matches won based on decision after toss
plt.figure(figsize=(20,10))
sns.countplot('season',hue='toss_decision',data=data,palette="summer")
plt.title("Number of Matches won by Toss Result",fontsize=24)
plt.xlabel('Season',fontsize=18)
plt.ylabel('Count',fontsize=18)
plt.show()
```

```python
#Winner per season
final_matches=data.drop_duplicates(subset=['season'],keep='last')
final_matches[['season','winner']].reset_index(drop=True).sort_values('season')
```

```python
#Winning chances
match=final_matches.win_by.value_counts()
labels=np.array(Toss.index)
sizes=match.values
colors=['#FFC1CC','#E7FEFF']
plt.figure(figsize=(15,10))
plt.pie(sizes,labels=labels, colors=colors,
       autopct='%1.1f%%',shadow=True, startangle=90)
plt.title('Match result', fontsize=26)
plt.axis('equal')
plt.show()
```

```python
Toss=final_matches.toss_decision.value_counts()
labels=np.array(Toss.index)
sizes=Toss.values
colors=['#FFFFC2','#964B00']
plt.figure(figsize=(15,10))
plt.pie(sizes,labels=labels, colors=colors,
       autopct='%1.1f%%',shadow=True, startangle=90)
plt.title('Toss result', fontsize=26)
plt.axis('equal')
plt.show()
```

```python
plt.figure(figsize=(20,15))
top_players=data.player_of_match.value_counts()[:10]
fig, ax=plt.subplots()
ax.set_ylim([0,20])
ax.set_ylabel("Count")
ax.set_title("Top player of the match winners")
top_players.plot.bar()
sns.barplot(x=top_players.index, y=top_players, orient='v', palette="coolwarm")
plt.show()
```

```python
#number of fours
four_data=complete_data[complete_data['batsman_runs']==4]
four_data.groupby('batting_team')['batsman_runs'].agg([('runs by fours','sum'),('fours','count')])
```

```python
#Number of fours
batsman_four=four_data.groupby('batsman')['batsman_runs'].agg([('four','count')]).reset_index().sort_values('four',ascending=0)

ax=batsman_four.iloc[:10,:].plot('batsman','four',kind='bar',color='pink')
plt.title("Number of fours hit",fontsize=24)
plt.xticks(rotation=30)
plt.xlabel('Player Name',fontsize=18)
plt.ylabel('Number of Fours',fontsize=18)
plt.show()
```

```python
#no of fours hit in each in each season
ax=four_data.groupby('season')['batsman_runs'].agg([('four','count')]).reset_index().plot('season','four',kind='bar',color="lightblue")
plt.title("Number of Fours in Each Season",fontsize=24)
plt.xticks(rotation=30)
plt.xlabel('Season',fontsize=18)
plt.ylabel('Number of Fours',fontsize=18)
plt.show()
```

```python
#number of sixes
six_data=complete_data[complete_data['batsman_runs']==6]
six_data.groupby('batting_team')['batsman_runs'].agg([('runs by sixes','sum'),('fours','count')])
```

```python
#Number of sixes
batsman_six=six_data.groupby('batsman')['batsman_runs'].agg([('sixes','count')]).reset_index().sort_values('sixes',ascending=0)
ax=batsman_six.iloc[:10,:].plot('batsman','sixes',kind='bar',color='black')
plt.title("Number of sixes hit",fontsize=24)
plt.xticks(rotation=30)
plt.xlabel('Player Name',fontsize=18)
plt.ylabel('Number of Sixes',fontsize=18)
plt.show()
```

```python
#no of Sixes hit in each in each season
ax=six_data.groupby('season')['batsman_runs'].agg([('sixes','count')]).reset_index().plot('season','sixes',kind='bar',color="orange")
plt.title("Number of Sixes in Each Season",fontsize=24)
plt.xticks(rotation=30)
plt.xlabel('Season',fontsize=18)
plt.ylabel('Number of Sixes',fontsize=18)
plt.show()
```

```python
#the top 10 leading run scorer in IPL
batsman_score=Data.groupby('batsman')['batsman_runs'].agg(['sum']).reset_index().sort_values('sum',ascending=False).reset_index(drop=True)
batsman_score=batsman_score.rename(columns={'sum':'batsman_runs'})
print("*** Top 10 Leading Run Scorer in IPL ***")
batsman_score.iloc[:10,:]
```

```python
# number of matches played by batsman
No_Matches_player= Data[["match_id","player_dismissed"]]
No_Matches_player=No_Matches_player .groupby("player_dismissed")["match_id"].count().reset_index().sort_values(by="match_id",ascending=False).reset_index(drop=True)
No_Matches_player.columns=["batsman","No_of Matches"]
No_Matches_player.head(5)
```

```python
# Dismissals in IPL
plt.figure(figsize=(18,10))
ax=sns.countplot(Data.dismissal_kind)
plt.title("Dismissals in IPL",fontsize=20)
plt.xlabel("Dismissals kind",fontsize=15)
plt.ylabel("count",fontsize=15)
plt.xticks(rotation=90)
plt.show()
```

```python
wicket_data=Data.dropna(subset=['dismissal_kind'])
wicket_data=wicket_data[~wicket_data['dismissal_kind'].isin(['run out','retired hurt','obstructing the field'])]

# purple cap winner
wicket_data.groupby('bowler')['dismissal_kind'].agg(['count']).reset_index().sort_values('count',ascending=False).reset_index(drop=True).iloc[:10,:]
```
